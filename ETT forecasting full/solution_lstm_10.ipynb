## Preprocessing

```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

df = pd.read_csv('ETTh2_full.csv')
df.head(10)
```

```python
# Converting the dt column to datetime object 
df['Date'] = [pd.to_datetime(x) for x in df['Date']]
df.head(10)
```

```python
df = df.sort_values('Date')
df.head(10)
```

```python
# Creating the data column
df['date'] = [x.date() for x in df['Date']]
df.head(7)
```

```python
df['date'] = [x.date() for x in df['Date']]
df['month'] = [x.month for x in df['Date']]
df['year'] = [x.year for x in df['Date']]
df['day'] = [x.day for x in df['Date']]
df['hour'] = [x.hour for x in df['Date']]
df.head(7)
```

```python
df['day_cos'] = [np.cos(x * (2 * np.pi / 24)) for x in df['hour']]
df['day_sin'] = [np.sin(x * (2 * np.pi / 24)) for x in df['hour']]
df.head(7)
```

```python
df['timestamp'] = [x.timestamp() for x in df['Date']]
df.head(7)
```

```python
# Seconds in day 
s = 24 * 60 * 60 
# Seconds in year 
year = (365.25) * s

df['year_cos'] = [np.cos((x) * (2 * np.pi / year)) for x in df['timestamp']]
df['year_sin'] = [np.sin((x) * (2 * np.pi / year)) for x in df['timestamp']]
df.head(7)
```

## Prepare data

```python
INPUT_SIZE = 48 # Dng `input_size` gi trc
LABEL_SIZE = 12  #  d on nhit  trong `label_size` gi
BATCH_SIZE = 32
FEATURE_FINAL = ['HUFL', 'HULL', 'MUFL', 'MULL', 'LUFL', 'OT', 'day_cos', 'day_sin', 'year_cos', 'year_sin']
label_name = 'OT'
```

```python
def slicing_window(df, df_start_idx, df_end_idx, input_size, label_size, label_name):
    features = [] # Khai bo list dng  lu tr cc X
    labels = []   # Khai bo list dng  lu tr cc y

    window_size = input_size + label_size # Tnh kch thc ca window

    # Nu df_end_idx = ch mc cui cng bng d liu, cn phi di xung 1 khong = window size 
    if df_end_idx == None:
        df_end_idx = len(df) - window_size 
    
    # Duyt qua tng mu d liu
    for idx in range(df_start_idx, df_end_idx):
        feature_end_idx = idx + input_size # Tnh ch mc kt thc ca X
        label_start_idx = idx + window_size - label_size # Tnh ch mc bt u ca y

        feature = df[idx:feature_end_idx] # Ly X
        label = df[label_name][label_start_idx:(idx+window_size)] # Ly y

        features.append(feature) 
        labels.append(label)

    # Chuyn list thnh np.ndarrray
    features = np.array(features)
    print(features.shape)
    labels = np.array(labels)

    return features, labels
```

```python
dataset_length = len(df) # S lng mu d liu trong DataFrame
TRAIN_SIZE = 0.8         # T l kch thc tp train
VAL_SIZE = 0.1           # T l kch thc tp val

# i t l thnh ch mc
TRAIN_END_IDX = int(TRAIN_SIZE * dataset_length) 
VAL_END_IDX = int(VAL_SIZE * dataset_length) + TRAIN_END_IDX
df = df[FEATURE_FINAL]

# Khi to X, y cho tp train
X_train, y_train = slicing_window(df, 
                                  df_start_idx=0,
                                  df_end_idx=TRAIN_END_IDX,
                                  input_size=INPUT_SIZE,
                                  label_size=LABEL_SIZE,
                                  label_name = label_name)

# Khi to X, y cho tp val
X_val, y_val = slicing_window(df, 
                              df_start_idx=TRAIN_END_IDX,
                              df_end_idx=VAL_END_IDX,
                              input_size=INPUT_SIZE,
                              label_size=LABEL_SIZE,
                              label_name = label_name)

# Khi to X, y cho tp test
X_test, y_test = slicing_window(df, 
                                df_start_idx=VAL_END_IDX,
                                df_end_idx=None,
                                input_size=INPUT_SIZE,
                                label_size=LABEL_SIZE,
                                label_name = label_name)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
import tensorflow as tf

# Khi to tf.data.Dataset
train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train)).batch(BATCH_SIZE)
val_ds   = tf.data.Dataset.from_tensor_slices((X_val, y_val)).batch(BATCH_SIZE)
test_ds  = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(BATCH_SIZE)

# Cu hnh cc tham s ti u cho vic c d liu
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds   = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

## Model and train

```python
# Khai bo normalization layer
normalize_layer = tf.keras.layers.Normalization(axis=-1)
normalize_layer.adapt(np.vstack((X_train, X_val, X_test)))
```

```python
input_shape = X_train.shape[-2:]
model = tf.keras.Sequential([tf.keras.Input(shape=input_shape, name='input_layer'), 
                             normalize_layer,  
                             tf.keras.layers.LSTM(32, return_sequences=True), 
                             tf.keras.layers.LSTM(32), 
                             tf.keras.layers.Dense(16),  
                             tf.keras.layers.Dense(LABEL_SIZE)])

model.summary()
```

```python
# Khai bo mt s gi tr siu tham s
EPOCHS = 30
LR = 1e-3

# Cu hnh mt s thng tin cho m hnh
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=LR),
              loss=tf.keras.losses.MeanAbsoluteError())
history = model.fit(train_ds, 
                    validation_data=val_ds,
                    epochs=EPOCHS)
```

```python
# nh gi m hnh trn tp test
test_evaluation = model.evaluate(test_ds)
```

```python
train_loss = history.history['loss']# c thng tin loss trn tp train
val_loss = history.history['val_loss'] # c thng tin loss trn tp val

plt.figure(figsize=(10, 5)) # Ci t kch thc khung nh

plt.subplot(1, 2, 1) # Khi to khung nh cho training loss
plt.xlabel('Epochs') # Hin th tn trc honh l 'Epochs'
plt.ylabel('Loss') # Hin th tn trc tung l 'Loss'
plt.title('Training loss') # Hin th title ca khung nh hin ti l 'Training Loss'
plt.plot(train_loss, color='red') # V ng gi tr loss trn tp train qua tng epoch (ng v mu )

plt.subplot(1, 2, 2) # Khi to khung nh cho val loss
plt.xlabel('Epochs') # Hin th tn trc honh l 'Epochs'
plt.ylabel('Loss') # Hin th tn trc tung l 'Loss'
plt.title('Validation loss') # Hin th title ca khung nh hin ti l 'Validation loss'
plt.plot(val_loss, color='orange') # V ng gi tr loss trn tp val qua tng epoch (ng v mu )

plt.show() # Hin th 2 khung nh nh
```
